```python
import urllib
import re
import csv
import os
import json

def pull_ids(urllink):

    htmlfile=urllib.urlopen(urllink)
    htmltext= htmlfile.read()
    regex= '<a href="/chaxun/list/(.+?).html">'
    pattern = re.compile(regex,re.DOTALL)
    number_list= re.findall(pattern, htmltext)

    return number_list

ids = {}

temp = []
temp += pull_ids("http://www.shicimingju.com/shicimark/tianyuanshi_1_0__1.html")
ids["tianyuan"] = temp

temp = []
temp += pull_ids("http://www.shicimingju.com/shicimark/sonbieshi_1_0__1.html")
ids["songbie"] = temp

temp = []
temp += pull_ids("http://www.shicimingju.com/shicimark/shanshuishi_1_0__1.html")
temp += pull_ids("http://www.shicimingju.com/shicimark/shanshuishi_2_0__1.html")
ids["shanshui"] = temp

temp = []
temp += pull_ids("http://www.shicimingju.com/shicimark/aiqingshi_1_0__1.html")
temp += pull_ids("http://www.shicimingju.com/shicimark/aiqingshi_2_0__1.html")
ids["aiqing"] = temp

temp = []
temp += pull_ids("http://www.shicimingju.com/shicimark/zhanzhengshi_1_0__1.html")
temp += pull_ids("http://www.shicimingju.com/shicimark/zhanzhengshi_2_0__1.html")
ids["zhanzheng"] = temp
```

```python
from bs4 import BeautifulSoup

def pull_poem(id_):
    htmlfile=urllib.urlopen("http://www.shicimingju.com/chaxun/list/{}.html".format(id_))
    htmltext= htmlfile.read()
    parsed_html = BeautifulSoup(htmltext, "html5lib")
    temp = parsed_html.body.find(class_="shici-content")
    return temp.get_text()

s = pull_poem("2973")
```
