# Task 1

PDF:
$ P(x) = \frac{\sin{x}}{2} \quad x \in [0, \pi]$
- one mode
- symmetric
- most of the probability mass will be at $\frac{\pi}{2}$
- There is no parameter; normalization constant is $\frac{1}{2}$


# Task 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf(x):
    return np.sin(x)/2

def cdf(x):
    '''int_0^x p(x)dx
    '''
    return (1-np.cos(x))/2

def ppf(q):
    '''inverse function of cdf
    '''
    return np.arccos(1-2*q)
```

```python
def plot_func(func, end):
    x = np.linspace(0,end,1000)
    plt.plot(x, globals()[func](x))
    plt.title(func)
    plt.show()
```

```python
plot_func('pdf', np.pi)
```

```python
plot_func('cdf', np.pi)
```

```python
plot_func('ppf', 1)
```

# Task 3

```python
def generate_samples(n):
    '''Sampling from uniform distribution, mapping to support
    '''
    return [ppf(np.random.random()) for i in range(n)]
```

```python
samples = generate_samples(100000)
plt.hist(samples, bins=50, density=True, alpha=0.5)
x = np.linspace(0, np.pi)
plt.plot(x, pdf(x))
plt.title("Histogram of random samples")
plt.xlabel("sample value")
plt.ylabel("frequency")
plt.show()
```
