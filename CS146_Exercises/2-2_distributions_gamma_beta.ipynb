```python
from scipy.stats import gamma, norm, beta
import numpy as np
import matplotlib.pyplot as plt
```

# Gamma

```python
x = np.linspace(0, 20, 1000)
y1 = gamma.pdf(x, a=5, scale=0.5)
y2 = gamma.pdf(x, a=9, scale=1.0)
```

```python
plt.plot(x, y1,label='a=9, theta=1')
plt.plot(x, y2,label='a=5, theta=0.5')
plt.legend()
plt.show()
```

```python
rand_samples = gamma.rvs(a=5, scale=0.5, size=100)
```

```python
plt.hist(rand_samples, density=True)
plt.plot(x, y1)
plt.show()
```

# Normal

```python
x = np.linspace(-5, 5, 1000)
y1 = norm.pdf(x, loc=0, scale=1)
y2 = norm.pdf(x, loc=1, scale=0.5)
```

```python
plt.plot(x, y1,label='mu=0, sigma=1')
plt.plot(x, y2,label='mu=1, sigma=0.5')
plt.legend()
plt.show()
```

```python
rand_samples = norm.rvs(loc=0, scale=1, size=100)
```

```python
plt.hist(rand_samples, density=True)
plt.plot(x, y1)
plt.show()
```

# Beta

```python
x = np.linspace(0, 1, 1000)
y1 = beta.pdf(x, 0.5,0.5)
y2 = beta.pdf(x, 2,2)
```

```python
plt.plot(x, y1,label='a=b=0.5')
plt.plot(x, y2,label='a=b=2')
plt.legend()
plt.show()
```

```python
rand_samples = beta.rvs(0.5, 0.5, size=100)
```

```python
plt.hist(rand_samples, density=True)
plt.plot(x, y1)
plt.show()
```
